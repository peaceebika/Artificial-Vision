```python
# Read image
img = cv2.imread("pepper.jpg")
```

```python
# Convert from BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Resize using bilinear interpolation
bilinear = cv2.resize(img_rgb, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)

# Resize using bicubic interpolation
bicubic = cv2.resize(img_rgb, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
```

```python
# Save output images
cv2.imwrite("bilinear_result.jpg", cv2.cvtColor(bilinear, cv2.COLOR_RGB2BGR))
cv2.imwrite("bicubic_result.jpg", cv2.cvtColor(bicubic, cv2.COLOR_RGB2BGR))
```

```python
# Display results
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Display Bilinear
plt.figure(figsize=(6,6))
plt.imshow(bilinear)
plt.title("Bilinear Interpolation")
plt.axis("off")
plt.show()
```

```python
# Display Bicubic
plt.figure(figsize=(6,6))
plt.imshow(bicubic)
plt.title("Bicubic Interpolation")
plt.axis("off")
plt.show()
```
